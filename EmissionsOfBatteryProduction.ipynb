```kotlin
val amountAsString by column<String>("Amount")
val amount by column<Int>()
val carbonFootprintPerKg by column<Double>()
val carbonFootprintTotal by column<Double>()
val mineral by column<String>("Mineral")

val totalPer60kwh = DataFrame.readCSV(Paths.get("raw-data").resolve("other").resolve("60-kwh-batery.tsv").inputStream())
    .add(amount) {
        amountAsString.getValue(it).replace("kg", "").toInt()
    }
    .add(carbonFootprintPerKg) {
        when (mineral.getValue(it)) {
            "Graphite" -> 5.31519
            "Aluminum" -> 15.1
            "Nickel" -> 13.0
            "Copper" -> 3.8
            "Steel" -> 1.85
            "Manganese" -> 6.0
            "Cobalt" -> 28.2
            "Lithium" -> 15.0
            "Iron" -> 2.0
            else -> throw IllegalStateException("Value was ${mineral.getValue(it)}")
        }
    }
    .add(carbonFootprintTotal) {
        carbonFootprintPerKg.getValue(it) * amount.getValue(it)
    }.sum(carbonFootprintTotal)

val perKwh = totalPer60kwh/60

println("CO2 generated on average per 1 KWh of capacity = ${perKwh} kg")
println("CO2 generated on Hummer EV batery : ${212.0 *perKwh}")
```

```kotlin
fun Pair<Double, Double>.average(): Double {
    return (this.first + this.second) / 2.0
}


val suvShare = listOf(Pair(58.0, 77.4).average(), Pair(72.0, 91.0).average(), 100.0).average() * (45.9 + 8.8)

val pickupShare = listOf(Pair(98.0, 131.0).average(), Pair(104.0, 141.0).average(), 123.0).average() * 19.3

val midsizeShare = listOf(Pair(57.5, 75.0).average(), Pair(62.0, 82.0).average(), 51.0).average() * (7.1 + 8.1)

val luxuryShare = listOf(112.0, 105.6, 118.0).average() * (5.3 + 0.7)

val vanShare = listOf(89.0,110.0,113.0).average() * 4.2

val weightedAverageBatteryCapacity = listOf(suvShare,pickupShare,midsizeShare,luxuryShare,vanShare).sum() / 100.0

println("US average car EV baterry should be $weightedAverageBatteryCapacity")
```

```kotlin
val gasCars = 288_500_000
val dieselCars = 9_900_000
val electricCars = 3_000_000
val chargingWithLosses = 1.13
val averageMileage = 13_476
val milesPerKwh = 4
val yearlyCharges = averageMileage / milesPerKwh / weightedAverageBatteryCapacity

val totalEnergyNeededPerYear =
    (weightedAverageBatteryCapacity * (gasCars + dieselCars) * chargingWithLosses * yearlyCharges) / 1000.0 
println("Total energy needed per year for all EVs : ${totalEnergyNeededPerYear.roundToLong()} MWh")
//1 135 999 848
val totalEnergyEVCurrentlyUsed =
    (weightedAverageBatteryCapacity * electricCars * chargingWithLosses * yearlyCharges) / 1000.0
```

```kotlin
val energyGeneratedYearly by column<Double>("GENNTAN")

val yearlyEnergy = DataFrame.readExcel(
    Paths.get("raw-data").resolve("epa").resolve("egrid2022_data.xlsx").inputStream(),
    "GEN22",
    1
).sum(energyGeneratedYearly)

val totalCapacity = DataFrame.readExcel(
    Paths.get("raw-data").resolve("epa").resolve("egrid2022_data.xlsx").inputStream(),
    "GEN22",
    1
).sum("NAMEPCAP")
    .toDouble() * 24.0 *365.0
    
println("How much power  - curr. EVs : ${totalEnergyEVCurrentlyUsed.roundToLong()} MWh")//   //11 420 910
println("How much power we produce   : ${yearlyEnergy.roundToLong()} MWh")                //4 240 140 549       
println("How much we could have:     : ${totalCapacity.roundToLong()} MWh")              //11 223 394 344
println("How much we need            : ${totalEnergyNeededPerYear.roundToLong()} MWh")//    1 135 999 848
println("How much X more power we need : ${(totalEnergyNeededPerYear / yearlyEnergy).roundToInt()}")
   
```
